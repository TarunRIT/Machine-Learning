```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv(r"D:\Tarun\Programs\Python_Programming\Machine_Learning\ML_Pre\Imbalanced_Data_credit.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['Class'].value_counts()
```

```python
legit = data[data['Class'] == 0]
fraud = data[data['Class'] == 1]
```

```python
#Undersampling

legitSample = legit.sample(500)
```

```python
newData = pd.concat((legitSample, fraud), axis = 0)
newData = newData.reset_index()
newData.drop(columns = "index", axis = 1, inplace = True)
newData.head()
```

```python
newData.info()
```

```python
X = newData.iloc[:, :-1]
Y = newData['Class']
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
#Model
model = LogisticRegression()
```

```python
model.fit(xTrain, yTrain)
```

```python
#Training Evaluation 
prediction = model.predict(xTrain)
accuracy = accuracy_score(yTrain, prediction)

print(accuracy)
```

```python
#Testing Evaluation 
prediction = model.predict(xTest)
accuracy = accuracy_score(yTest, prediction)

print(accuracy)
```
