```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
```

```python
#Load Data and Preprocessing

data = pd.read_csv("House_Price_Prediction.csv")
data.columns.values[-1] = "price"
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
#Understanding Data
correlation = data.corr()

plt.figure(figsize = (10, 10)), 
sns.heatmap(correlation, cbar = True, square = True, fmt = ".1f", annot = True, annot_kws = {'size': 8}, cmap = "coolwarm" )
plt.show()
```

```python
#Features and Traget
X = data.iloc[:, :-1]
Y = data.iloc[:, -1]
```

```python
#Train Test Split
xTrain, xTest, yTrain, yTest = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
#Model
model = XGBRegressor(random_state = 2)
model.fit(xTrain, yTrain)
```

```python
#%pip install --upgrade scikit-learn xgboost
```

```python
#Training Evaluation
predictions = model.predict(xTrain)

score1 = metrics.r2_score(yTrain, predictions)
score2 = metrics.mean_absolute_error(yTrain, predictions)

print("R2 Error: ", score1)
print("MAE: ", score2)
```

```python
#Visualize Training Evaluation
plt.scatter(yTrain, predictions, color = "Blue")
plt.title("Actual Price Vs Predicted Price")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```

```python
#Model Evaluation
predictions = model.predict(xTest)

score1 = metrics.r2_score(yTest, predictions)
score2 = metrics.mean_absolute_error(yTest, predictions)

print("R2 Error: ", score1)
print("MAE: ", score2)
```

```python
#Visualize Model Evaluation
plt.scatter(yTest, predictions, color = "Red")
plt.title("Actual Price Vs Predicted Price")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```
